```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import sys, os
%matplotlib inline
```

```python
# Make a list of calibration images
Images = [cv2.imread(filename) for filename in os.listdir('./')]
fname = 'calibration_test.png'
img = cv2.imread(fname)
```

```python
print('Raw Calibration Image Shape: ', img.shape)
```

```python
# Array to store object points and image points from all images
objPoints = [] # 3D points in real world
imgPoints = [] # 2D points in image plane
```

```python
# Display raw image
plt.imshow(img)
plt.show()
```

```python
# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print('Gray Calibration Image Shape: ', gray.shape)
plt.imshow(gray, cmap='gray')
plt.show()
```

```python
# prepare object points
nx = 8 # number of inside corners in x
ny = 6 # number of inside corners in y
```

```python
# Find the chessboard corners
ret, corners = cv2.findChessboardCorners(gray, (nx, ny), None)
print('Found all corners: ', ret)
print('No of Corners: ', corners.shape)
```

```python
# If found, draw corners
if ret == True:
    # Draw and display the corners
    cv2.drawChessboardCorners(img, (nx, ny), corners, ret)
    plt.imshow(img)
    plt.show()
```
